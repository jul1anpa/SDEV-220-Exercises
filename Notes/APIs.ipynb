**An API is an Application Programming Interface**

An API allows two applications to interact with each other even if they are built using different languages.

An API involves a server and a client with a server giving responses to client requests and the clients consuming the information.

JSON, or JavaScript Object Notation, is the standard communication protocol for applications.

REST, or representational state transfer, is the means of communication for an API.

An API endpoint is the entry point or address that an application will provide that enables interaction with other software applications.

There are various methods of requesting data from a server.

These are called HTTP methods and they include:

**1. GET -- Which retrieves data**

**2. POST -- Which writes new data**

**3. DELETE -- Which deletes data**

**4. PUT -- Which writes or updates data**

CRUD is an acronym that represents how you should be able to interact with a database.

**C -- Create data**

**R -- Read data**

**U -- Update data**

**D -- Delete data**

```python
import requests
import json

response = requests.get('https://api.stackexchange.com/2.3/questions?order=desc&sort=activity&site=stackoverflow')

for question in response.json()['items']:
    if question['answer_count'] == 0:
        print(question['title'])
        print(question['link'])
        print()
    else:
        print("Skipped")
    print()
```
